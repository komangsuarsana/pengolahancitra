```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("bedugul.jpeg", 1)
```

```python
half = cv2.resize(image, (0,0), fx = 0.1, fy=0.1)
bigger = cv2.resize(image, (1050, 1610))
```

```python
 
stretch_near = cv2.resize(image, (780, 540),
               interpolation = cv2.INTER_NEAREST)
```

```python
titles = ["Original", "Half", "Bigger", "Interpolation Nearest"]
```

```python
images = [image, half, bigger, stretch_near]
```

```python
count  = 4 
for i in range(count):
    plt.subplot(2, 2, i + 1)
    plt.title(titles[i])
    plt.imshow(images[i])
```

```python
gaussian =  cv2.GaussianBlur(image, (7,7), 0)
median = cv2.medianBlur(image, 7)
bilateral = cv2.bilateralFilter(image, 7, 75,75)

total = 4
imageTitle = ["Gaussian", "median", "bilateral", "original"]
imageFilter = [gaussian, median, bilateral, image]
for i in range(total):
    plt.subplot(2, 2, i + 1)
    plt.title(imageTitle[i])
    plt.imshow(imageFilter[i])
```
