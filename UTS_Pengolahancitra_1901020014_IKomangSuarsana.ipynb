```python
#Soal no 2A
import cv2
import matplotlib.pyplot as plt
from PIL import Image
import requests
```

```python
image = cv2.imread("jalakbali.jpg", 1)
```

```python
image1 = cv2.resize(image, (1024, 1024))
image2 = cv2.resize(image, (600, 600))
```

```python
titles = ["ImgOriginal", "1024x1024", "600*600"]
images = [image, image1, image2]
```

```python
count  = 3
for i in range(count):
    plt.subplot(2, 2, 1 + i)
    plt.title(titles[i])
    plt.imshow(images[i])
```

```python
#Soal no 2B
median = cv2.medianBlur(image, 7)
```

```python
plt.imshow(image)
```

```python
count = 2
imageTitle = ["ImgOriginal", "Median"]
imageFilter = [image, median]
for i in range(count):
    plt.subplot(2, 2, 1 + i)
    plt.title(imageTitle[i])
    plt.imshow(imageFilter[i])
```

```python
blur_1024 = cv2.cvtColor(cv2.medianBlur(image1, 11), cv2.COLOR_BGR2GRAY)
blur_600 = cv2.cvtColor(cv2.medianBlur(image2, 11), cv2.COLOR_BGR2GRAY)

plt.subplot(121)
plt.imshow(blur_1024, cmap='gray')
plt.subplot(122)
plt.imshow(blur_600, cmap='gray');
```
