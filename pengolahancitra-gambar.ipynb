```python
from PIL import Image
import requests
```

```python
url = "https://www.ayobali.net/wp-content/uploads/2020/09/Tempat-Wisata-Bedugul-Bali.jpg"
gambar = Image.open(requests.get(url, stream=True).raw)
```

```python
gambar
```

```python
imageGray = gambar.convert('1')
imageGray
```

```python
import numpy as np
```

```python
checkImage = np.array(gambar)
checkImage
```

```python
print(checkImage.shape)
```

```python
inversion = 255 - checkImage
```

```python
visualImage = Image.fromarray(inversion, 'RGB')
```

```python
visualImage
```

```python
im_32 = checkImage // 32*32
im_128 = checkImage // 128*128


imageConcat = np.concatenate((checkImage, im_32, im_128), axis=1)

visualImage = Image.fromarray(imageConcat, 'RGB')
```
