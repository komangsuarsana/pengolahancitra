```python
import cv2
import matplotlib.pyplot as plt
```

```python


image = cv2.imread('image1.jpeg')

img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

val, treshBinary = cv2.threshold(img, 150,250, cv2.THRESH_BINARY);
val, treshBinaryInv = cv2.threshold(img, 120,120, cv2.THRESH_BINARY_INV);
val, treshTrunc = cv2.threshold(img, 120,120, cv2.THRESH_TRUNC);
val, treshToZero = cv2.threshold(img, 120,120, cv2.THRESH_TOZERO);
val, treshTozeroInv = cv2.threshold(img, 120,120, cv2.THRESH_TOZERO_INV);

title = ["Binary", "Binary Inverted", "Truncate", "To Zero", "To Zero inverted"];
images = [treshBinary, treshBinaryInv, treshTrunc, treshToZero, treshTozeroInv]

count = 5 ;



for i in range(count) : 
  plt.subplot(3,2,1 + i)
  plt.title(title[i])
  plt.imshow(images[i])

plt.show()
```
